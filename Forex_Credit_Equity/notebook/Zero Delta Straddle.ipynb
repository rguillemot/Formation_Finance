```python
import numpy as np
from scipy.stats import norm
import math

def BSFX(S,K,T,cp,rd,rf,vol):
    DFd = exp(-rd*T)
    DFf = exp(-rf*T)
    F = S*DFf/DFd
    if vol*T == 0:
        return max(cp*DFd*(F-K),0)
    else:
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        d2 = d1 - vol*math.sqrt(T)
        return (cp*F*norm.cdf(cp*d1)-cp*K*norm.cdf(cp*d2))*DFd

def BSDelta(S,K,T,cp,rd,rf,vol):
    DFd = exp(-rd*T)
    DFf = exp(-rf*T)
    if vol*T == 0:
        if cp*S > cp*K:
            return cp*DFf
        else:
            return 0.0
    else:
        F = S*DFf/DFd
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        return cp*norm.cdf(cp*d1)*DFf
```

```python
NEUR = 100000000
NUSD = 139750000

S = 1.3889
rUSD = 0.003
rEUR =0.005
m=0.001
vol = 0.12
K = float(NUSD)/NEUR
T = 1.0

KMin = 1.35
KMax = 1.45
PMin = 0
PMax = 17000000

def xfunc(x, pos=0):
         return '%1.2f'%(x)
    
def yfunc(y, pos=0):
     return '%1.0f Mios'%(y/1e6)
    
def yfunc_delta(y, pos=0):
         return '%1.0f'%(y*100)+'\%'


def format_graph(ax,x,y,KMin,KMax,PMin,PMax,title,xtitle,ytitle,color,delta=False):
    ax.set_xlim([0.99*KMin,1.01*KMax])
    ax.set_ylim([PMin,PMax])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    ax.spines['bottom'].set_position('zero')
    ax.xaxis.set_major_formatter(FuncFormatter(xfunc))
    if delta:
        ax.yaxis.set_major_formatter(FuncFormatter(yfunc_delta))
    else:
        ax.yaxis.set_major_formatter(FuncFormatter(yfunc))
    ax.set_title(title,fontsize=20)
    ax.set_xlabel(xtitle)
    ax.set_ylabel(ytitle)
    ax.plot(x,y,color=color,linewidth=3)
    
def delta_EUR_USD(K):
    delta = BSDelta(S,K,T,1,rUSD,rEUR-m,vol)+BSDelta(S,K,T,-1,rUSD,rEUR-m,vol)
    return delta
    
def zero_delta_straddle(filename,slide):
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(16,5))
    
    EURUSD = np.linspace(KMin, KMax, 100)
    payoff=[max(NEUR*s-NUSD,0)+max(NUSD-NEUR*s,0) for s in EURUSD]
    value=[BSFX(s,K,T,1,rUSD,rEUR-m,vol)*NEUR+BSFX(s,K,T,-1,rUSD,rEUR-m,vol)*NEUR for s in EURUSD]
    delta=[delta_EUR_USD(s) for s in EURUSD]
    
    if slide>=1:
        format_graph(ax1,EURUSD,payoff,KMin,KMax,PMin,PMax,u'Payoff et valeur','F (EUR/USD)','USD','black')
    else:
        format_graph(ax1,[],[],KMin,KMax,PMin,PMax,u'Payoff et valeur','F (EUR/USD)','USD','black')
    if slide>=2:
        format_graph(ax1,EURUSD,value,KMin,KMax,PMin,PMax,u'Payoff et valeur','F (EUR/USD)','USD','blue')
    else:
        format_graph(ax1,[],[],KMin,KMax,PMin,PMax,u'Payoff et valeur','F (EUR/USD)','USD','blue')
        
    if slide>=3:
        format_graph(ax2,EURUSD,delta,KMin,KMax,-0.25,0.25,u'$\delta$ = f(K)','K (EUR/USD)',u'\%EUR','blue',True)
    else:
        format_graph(ax2,[],[],KMin,KMax,-0.25,0.25,u'$\delta$ = f(K)','K (EUR/USD)',u'\%EUR','blue',True)
    F=exp((rUSD-rEUR)*T)*S
    if slide>=4:
        ax2.scatter(F,delta_EUR_USD(F),s=150, facecolors='none', edgecolors='black')
        ax2.annotate(u'Pour K=F $\delta$=4\%', xy=(F+0.003, delta_EUR_USD(F)+0.02),  xycoords='data',
                    xytext=(+20, 40), textcoords='offset points',size=20,
                    arrowprops=dict(arrowstyle="->",facecolor='black'),
                    horizontalalignment='left', verticalalignment='bottom',
                    )
    if slide>=5:
        ax2.annotate(u'Que vaut $K^{ATM}$ ?', xy=(K+0.003, delta_EUR_USD(K)+0.02),  xycoords='data',
                    xytext=(+20, 30), textcoords='offset points',size=20,
                    arrowprops=dict(arrowstyle="->",facecolor='black'),
                    horizontalalignment='left', verticalalignment='bottom',
                    )
        
    plt.savefig(filename, bbox_inches='tight')
```
