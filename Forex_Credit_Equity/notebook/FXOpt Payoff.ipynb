```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
rcParams['text.latex.unicode']=True
```

```python
NEUR = 100000000
NUSD = 139000000

KMin = 1.15
KMax = 2.05
PMin = -10000000
PMax = 50000000

def xfunc(x, pos=0):
         return '%1.2f'%(x)
def yfunc(y, pos=0):
     return '%1.0f Mios'%(y/1e6)


def format_graph(ax,x,y,KMin,KMax,PMin,PMax,title,xtitle,ytitle,color):
    ax.set_xlim([KMin,KMax])
    ax.set_ylim([PMin,PMax])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    ax.spines['bottom'].set_position('zero')
    ax.xaxis.set_major_formatter(FuncFormatter(xfunc))
    ax.yaxis.set_major_formatter(FuncFormatter(yfunc))
    ax.set_title(title)
    ax.set_xlabel(xtitle)
    ax.set_ylabel(ytitle)
    ax.plot(x,y,color=color,linewidth=3)


def payoff(filename,slide):
    fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(10,5))
    
    EURUSD = np.linspace(KMin, KMax, 100)
    USDEUR = np.linspace(1.0/KMax, 1.0/KMin, 100)
    EURUSD_USD=[max(NEUR*s-NUSD,0) for s in EURUSD]
    EURUSD_EUR=[max(NEUR*s-NUSD,0)/s for s in EURUSD]
    USDEUR_EUR=[max(NEUR-NUSD*s,0) for s in USDEUR]
    USDEUR_USD=[max(NEUR-NUSD*s,0)/s for s in USDEUR]
    
    if slide >= 1:
        format_graph(ax3,EURUSD,EURUSD_USD,KMin,KMax,PMin,PMax,'EUR/USD - USD','EUR/USD','USD','blue')
    else:
        format_graph(ax3,[],[],KMin,KMax,PMin,PMax,'EUR/USD - USD','EUR/USD','USD','blue')
    if slide >= 2:
        format_graph(ax1,EURUSD,EURUSD_EUR,KMin,KMax,PMin,PMax,'EUR/USD - EUR','EUR/USD','EUR','red')
    else:
        format_graph(ax1,[],[],KMin,KMax,PMin,PMax,'EUR/USD - EUR','EUR/USD','EUR','red')
    if slide >= 3:
        format_graph(ax2,USDEUR,USDEUR_EUR,1.0/KMax,1.0/KMin,PMin,PMax,'USD/EUR - EUR','USD/EUR','EUR','red')
    else:
        format_graph(ax2,[],[],1.0/KMax,1.0/KMin,PMin,PMax,'USD/EUR - EUR','USD/EUR','EUR','red')
    if slide >= 4:
        format_graph(ax4,USDEUR,USDEUR_USD,1.0/KMax,1.0/KMin,PMin,PMax,'USD/EUR - USD','USD/EUR','USD','blue')
    else:
        format_graph(ax4,[],[],1.0/KMax,1.0/KMin,PMin,PMax,'USD/EUR - USD','USD/EUR','USD','blue')
    plt.savefig(filename, bbox_inches='tight')
```
