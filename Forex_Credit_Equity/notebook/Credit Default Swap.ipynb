```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plotFixFlow(ax,starty,t,value,color,label):
    ax.arrow(t,starty,0,value,color=color,head_width=0.05, head_length=0.1,linewidth=0.4)
    ax.annotate(label, xy=(t,starty),  
                xycoords='data', size=20, xytext=(10, 25), textcoords='offset points'
                )


def plotVarFlow(ax,starty,t,value,color,label):
    y = starty + np.arange(0,value+0.005,0.005)
    x = [t+np.sin(v/value*np.pi*10)*0.02 for v in y]
    ax.plot(x,y,color='black',linewidth=0.4)
    ax.arrow(t,starty+value+0.02,0,0.0001,color=color,head_width=0.05, head_length=0.1,linewidth=2.5)
    ax.annotate(label, xy=(t,starty),  
                xycoords='data', size=20, xytext=(10, -25), textcoords='offset points'
                )
    
def plotdelta(ax,starty,t1,t2,color,label,coeff=1):
    x = np.arange(t1,t2+0.02,0.02)
    y = starty-(x-t1)*(x-t2)/(t1-t2)/(t1-t2)*0.3*coeff
    ax.plot(x,y,color=color,linewidth=0.4,linestyle='--')
    ax.annotate(label, xy=((t1+t2)/2,starty),  
                xycoords='data', size=20, xytext=(-10, coeff*15), textcoords='offset points'
                )


fig, ax = plt.subplots(figsize=(15,7.5))

ax.set_xlim([-0.8,3])
ax.set_ylim([-1.2,2])

plt.axis('off')

ax.arrow(0,0.9,0,0.2,color='black',linewidth=0.1)
ax.annotate('$T_0=t$', xy=(0,1),  
                xycoords='data', size=20, xytext=(-30, -25), textcoords='offset points'
                )

ax.arrow(2,0.9,0,0.2,color='black',linewidth=0.1)
ax.annotate('$T_n$', xy=(2,1),  
                xycoords='data', size=20, xytext=(+10, -25), textcoords='offset points'
                )

ax.arrow(1.5,0.9,0,0.2,color='black',linewidth=0.1)
ax.annotate('$\\tau$', xy=(1.5,1),  
                xycoords='data', size=20, xytext=(10, +25), textcoords='offset points'
                )


ax.arrow(0,-0.1,0,0.2,color='black',linewidth=0.1)
ax.annotate('$t$', xy=(0,0),  
                xycoords='data', size=20, xytext=(-2, -25), textcoords='offset points'
                )

ax.arrow(2,-0.1,0,0.2,color='black',linewidth=0.1)
ax.annotate('$T_n$', xy=(2,0),  
                xycoords='data', size=20, xytext=(+10, 25), textcoords='offset points'
                )

ax.arrow(-0.5,1,3,0,color='black',head_width=0.2, head_length=0.1,linewidth=0.1)

ax.arrow(-0.5,0,3,0,color='black',head_width=0.2, head_length=0.1,linewidth=0.1)

plotVarFlow(ax,1,0.3,0.4,'black','$T_1$')
plotVarFlow(ax,1,0.6,0.4,'black','$T_2$')
plotVarFlow(ax,1,0.9,0.4,'black','$T_3$')

plotdelta(ax,1,0,0.3,'black','$\delta_1$')
plotdelta(ax,1,0.3,0.6,'black','$\delta_2$')
plotdelta(ax,1,0.6,0.9,'black','$\delta_3$')
plotdelta(ax,1,0,1.5,'black','',-4)

plotFixFlow(ax,0,1.5,-0.6,'black','$\\tau$')
plotdelta(ax,0,0,1.5,'black','',4)

ax.annotate('Jambe de prime.', xy=(-0.8,1.2),  
                xycoords='data', size=17, xytext=(0, 10), textcoords='offset points'
                )

ax.annotate('Jambe de protection.', xy=(-0.8,0.2),  
                xycoords='data', size=17, xytext=(0, 10), textcoords='offset points'
                )

ax.annotate('$N \\times C$', xy=(1,1.2),  
                xycoords='data', size=17, xytext=(0, 10), textcoords='offset points'
                )

ax.annotate('$N \\times (1-Rec)$', xy=(1.6,-0.5),  
                xycoords='data', size=17, xytext=(0, 10), textcoords='offset points'
                )



plt.savefig('../figures/cds.png', bbox_inches='tight')
```
