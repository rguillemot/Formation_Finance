```python
import numpy as np
from scipy.stats import norm
import math

def BSFX(S,K,T,cp,rd,rf,vol):
    DFd = exp(-rd*T)
    DFf = exp(-rf*T)
    F = S*DFf/DFd
    if vol*T == 0:
        return max(cp*DFd*(F-K),0)
    else:
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        d2 = d1 - vol*math.sqrt(T)
        return (cp*F*norm.cdf(cp*d1)-cp*K*norm.cdf(cp*d2))*DFd

def BSDelta(S,K,T,cp,rd,rf,vol):
    DFd = exp(-rd*T)
    DFf = exp(-rf*T)
    if vol*T == 0:
        if cp*S > cp*K:
            return cp*DFf
        else:
            return 0.0
    else:
        F = S*DFf/DFd
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        return cp*norm.cdf(cp*d1)*DFf
```

```python
NEUR = 100000000
NUSDCall = 149000000
NUSDPut = 129000000
KCall = float(NUSDCall)/NEUR
KPut = float(NUSDPut)/NEUR

KMin = 1.2
KMax = 1.6
PMin = -12000000
PMax = 12000000

def xfunc(x, pos=0):
         return '%1.2f'%(x)
    
def yfunc(y, pos=0):
     return '%1.0f Mios'%(y/1e6)
    
def yfunc_delta(y, pos=0):
         return '%1.0f'%(y*100)+'\%'


def format_graph(ax,x,y,KMin,KMax,PMin,PMax,title,xtitle,ytitle,color,delta=False):
    ax.set_xlim([0.99*KMin,1.01*KMax])
    ax.set_ylim([PMin,PMax])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    ax.spines['bottom'].set_position('zero')
    ax.xaxis.set_major_formatter(FuncFormatter(xfunc))
    if delta:
        ax.yaxis.set_major_formatter(FuncFormatter(yfunc_delta))
    else:
        ax.yaxis.set_major_formatter(FuncFormatter(yfunc))
    ax.set_title(title,fontsize=20,color=color)
    ax.set_xlabel(xtitle)
    ax.set_ylabel(ytitle)
    ax.plot(x,y,color=color,linewidth=3)
    
def delta_EUR_USD(K):
    delta = BSDelta(S,K,T,1,rUSD,rEUR-m,vol)+BSDelta(S,K,T,-1,rUSD,rEUR-m,vol)
    return delta
    
def risk_reversal(filename,solution=False):
    fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(14,7))
    
    EURUSD = np.linspace(KMin, KMax, 100)
    payoff1=[max(NEUR*s-NUSDCall,0)-max(NUSDPut-NEUR*s,0) for s in EURUSD]
    payoff2=[-max(NEUR*s-NUSDCall,0)+max(NUSDPut-NEUR*s,0) for s in EURUSD]
    payoff3=[max(NEUR*s-NUSDCall,0)+max(NUSDPut-NEUR*s,0) for s in EURUSD]
    payoff4=[max(NEUR*s-NUSDPut,0)-max(NEUR*s-NUSDCall,0)-NEUR/10 for s in EURUSD]
    
    if solution:
        format_graph(ax1,EURUSD,payoff1,KMin,KMax,PMin,PMax,u'Solution 1','F (EUR/USD)','USD','green')
    else:
        format_graph(ax1,EURUSD,payoff1,KMin,KMax,PMin,PMax,u'Solution 1','F (EUR/USD)','USD','black')
    ax1.annotate(u'1.49', xy=(KCall+0.01, -1300000),  xycoords='data',
                xytext=(+20, -40), textcoords='offset points',size=20,
                arrowprops=dict(arrowstyle="->",facecolor='black'),
                horizontalalignment='left', verticalalignment='bottom',
                )
    ax1.annotate(u'1.29', xy=(KPut-0.01, +300000),  xycoords='data',
                xytext=(-20, +40), textcoords='offset points',size=20,
                arrowprops=dict(arrowstyle="->",facecolor='black'),
                horizontalalignment='right', verticalalignment='bottom',
                )
    
    if solution:
        format_graph(ax2,EURUSD,payoff2,KMin,KMax,PMin,PMax,u'Solution 2','F (EUR/USD)','USD','red')
    else:
        format_graph(ax2,EURUSD,payoff2,KMin,KMax,PMin,PMax,u'Solution 2','F (EUR/USD)','USD','black')
    ax2.annotate(u'1.49', xy=(KCall+0.01, 300000),  xycoords='data',
                xytext=(+20, +40), textcoords='offset points',size=20,
                arrowprops=dict(arrowstyle="->",facecolor='black'),
                horizontalalignment='left', verticalalignment='bottom',
                )
    ax2.annotate(u'1.29', xy=(KPut-0.01, -300000),  xycoords='data',
                xytext=(-20, -40), textcoords='offset points',size=20,
                arrowprops=dict(arrowstyle="->",facecolor='black'),
                horizontalalignment='right', verticalalignment='bottom',
                )
    
    if solution:
        format_graph(ax3,EURUSD,payoff3,KMin,KMax,PMin,PMax,u'Solution 3','F (EUR/USD)','USD','red')
    else:
        format_graph(ax3,EURUSD,payoff3,KMin,KMax,PMin,PMax,u'Solution 3','F (EUR/USD)','USD','black')
    ax3.annotate(u'1.49', xy=(KCall+0.01, -1300000),  xycoords='data',
                xytext=(+20, -40), textcoords='offset points',size=20,
                arrowprops=dict(arrowstyle="->",facecolor='black'),
                horizontalalignment='left', verticalalignment='bottom',
                )
    ax3.annotate(u'1.29', xy=(KPut-0.01, -300000),  xycoords='data',
                xytext=(-20, -40), textcoords='offset points',size=20,
                arrowprops=dict(arrowstyle="->",facecolor='black'),
                horizontalalignment='right', verticalalignment='bottom',
                )
    
    if solution:
        format_graph(ax4,EURUSD,payoff4,KMin,KMax,PMin,PMax,u'Solution 4','F (EUR/USD)','USD','red')
    else:
        format_graph(ax4,EURUSD,payoff4,KMin,KMax,PMin,PMax,u'Solution 4','F (EUR/USD)','USD','black')
    ax4.annotate(u'1.49', xy=(KCall+0.01, -1300000),  xycoords='data',
                xytext=(+20, -40), textcoords='offset points',size=20,
                arrowprops=dict(arrowstyle="->",facecolor='black'),
                horizontalalignment='left', verticalalignment='bottom',
                )
    ax4.annotate(u'1.29', xy=(KPut-0.01, -300000),  xycoords='data',
                xytext=(-20, -40), textcoords='offset points',size=20,
                arrowprops=dict(arrowstyle="->",facecolor='black'),
                horizontalalignment='right', verticalalignment='bottom',
                ) 
        
    plt.savefig(filename, bbox_inches='tight')
```
