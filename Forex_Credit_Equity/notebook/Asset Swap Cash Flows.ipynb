```python
%%tikz --scale 2 --size 1000,1000 -f png
\usetikzlibrary{shapes.misc,matrix,positioning,fit,decorations.pathmorphing}
\tikzset{cancel/.style={
        cross out,
        draw,
        red,
        ultra thick,
        fit=#1,
        },
    wave/.style={
        decorate,decoration={snake,post length=0.1cm, pre length=0.1cm,amplitude=1mm,
        segment length=0.5cm},thick}
}

\tikzstyle{cpty} = [draw,  minimum height=3em, minimum width=6em, rounded corners]
\node[cpty] (svc1) {Acheteur};
\path[-latex] 
    node[cpty, right of=svc1,xshift=4cm] (svc2) {Vendeur} (svc1.10) 
    edge[wave] node[midway,above] {LIBOR +x bps} (svc2.170);
\path[latex-] 
    node[cpty, below of=svc2,yshift=-1cm] (svc3) {Bond} (svc2) 
    edge node[midway,right] (svc4) {Coupon} (svc3);
\path[latex-] 
    (svc1.-10) 
    edge node[midway,below] {Coupon} 
    (svc2.190);
\node [cancel=(svc3)] {};
\node [cancel=(svc4)] {};
```
