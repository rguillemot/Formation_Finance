```python
%%tikz --scale 2 --size 1000,1000 -f png
\usetikzlibrary{shapes.misc,matrix,positioning,fit,decorations.pathmorphing}
\tikzset{cancel/.style={
        cross out,
        draw,
        red,
        ultra thick,
        fit=#1,
        },
    wave/.style={
        decorate,decoration={snake,post length=0.1cm, pre length=0.1cm,amplitude=1mm,
        segment length=0.5cm},thick}
}

\tikzstyle{cpty} = [draw,  minimum height=3em, minimum width=6em, rounded corners]
\node[cpty] (acheteur) {Acheteur};
\path[-latex] 
    node[cpty, right of=acheteur,xshift=4cm] (vendeur) {Vendeur} (acheteur.10) 
    edge[wave] node[midway,above] {\small{LIBOR +x bps}} (vendeur.170);
\path[latex-] 
    node[cpty, below of=vendeur,yshift=-1cm] (bond) {Bond} (vendeur) 
    edge node[midway,right] (coupon) {\small{Coupon}} (bond);
\path[latex-] 
    (acheteur.-10) 
    edge node[midway,below] {Coupon} 
    (vendeur.190);
\node [cancel=(bond)] {};
\node [cancel=(coupon)] {};
```
