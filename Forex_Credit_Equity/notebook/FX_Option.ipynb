```python
import numpy as np
from scipy.stats import norm
import math

def BSFX(S,K,T,cp,rd,rf,vol):
    if vol*T == 0:
        return cp*(F-K)
    else:
        F = S*exp((rd-rf)*T)
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        d2 = d1 - vol*math.sqrt(T)
    return (cp*F*norm.cdf(cp*d1)-cp*K*norm.cdf(cp*d2))*exp(-rd*(T))

def DeltaFX(S,K,T,cp,rd,rf,vol,delta_ccy,prem_ccy):
    eps = 1e-6
    premium = BSFX(S+eps,K,T,cp,rd,rf,vol)
    delta_raw = (premium-BSFX(S,K,T,cp,rd,rf,vol))/eps
    premium = premium/S
    if delta_ccy == 1 and prem_ccy == 1:
        return delta_raw - premium
    elif delta_ccy == 1 and prem_ccy == 2:
        return delta_raw
    elif delta_ccy == 2 and prem_ccy == 1:
        return -(delta_raw-premium)*S/K
    elif delta_ccy == 2 and prem_ccy == 2:
        return -(delta_raw)*S/K
    

def StraddleDeltaFX(S,K,T,rd,rf,vol,delta_ccy,prem_ccy):
    return DeltaFX(S,K,T,1,rd,rf,vol,delta_ccy,prem_ccy)+DeltaFX(S,K,T,-1,rd,rf,vol,delta_ccy,prem_ccy)

def ImpliedStrikeZeroDeltaStraddle(S,T,rd,rf,vol,inc_prem):
    K = S*exp((rd-rf)*T)
    eps = 1e-8
    for i in range(20):
        if inc_prem:
            val = StraddleDeltaFX(S,K,T,rd,rf,vol,1,1)
        else:
            val = StraddleDeltaFX(S,K,T,rd,rf,vol,1,2)
            
        if abs(val) < eps:
            return K
        
        if inc_prem:
            der = (StraddleDeltaFX(S,K+eps,T,rd,rf,vol,1,1)-val)/eps
        else:
            der = (StraddleDeltaFX(S,K+eps,T,rd,rf,vol,1,2)-val)/eps
            
        K = K-val/der
    
    return K

def ImpliedStrikeDelta(S,T,delta,rd,rf,vol,inc_prem):
    K = S*exp((rd-rf)*T)
    eps = 1e-8
    if delta > 0:
        cp = 1
    else:
        cp = -1
    for i in range(20):
        if inc_prem:
            val = DeltaFX(S,K,T,cp,rd,rf,vol,1,1)
        else:
            val = DeltaFX(S,K,T,cp,rd,rf,vol,1,2)
            
        if abs(val-delta) < eps:
            return K
        
        if inc_prem:
            der = (DeltaFX(S,K+eps,T,cp,rd,rf,vol,1,1)-val)/eps
        else:
            der = (DeltaFX(S,K+eps,T,cp,rd,rf,vol,1,2)-val)/eps
            
        K = K-(val-delta)/der
    
    return -1


def ImpliedVol(S,T,K,cp,rd,rf,price):
    vol = 0.15
    eps = 1e-8
    
    for i in range(20):
        val = BSFX(S,K,T,cp,rd,rf,vol)
            
        if abs(val-price) < eps:
            return vol
    
        der = (BSFX(S,K,T,cp,rd,rf,vol+eps)-val)/eps
            
        vol = vol-(val-price)/der
    
    return -1


def SmileFX(S,T,rd,rf,ATM,RR25,RR10,BF25,BF10,inc_prem):
    vol25Call = ATM + BF25 + 0.5*RR25
    vol25Put = ATM + BF25 - 0.5*RR25
    vol10Call = ATM + BF10 + 0.5*RR10
    vol10Put = ATM + BF10 - 0.5*RR10
    KATM = ImpliedStrikeZeroDeltaStraddle(S,T,rd,rf,ATM,inc_prem)
    K25Call = ImpliedStrikeDelta(S,T,0.25,rd,rf,vol25Call,inc_prem)
    K25Put = ImpliedStrikeDelta(S,T,-0.25,rd,rf,vol25Put,inc_prem)
    K10Call = ImpliedStrikeDelta(S,T,0.10,rd,rf,vol10Call,inc_prem)
    K10Put = ImpliedStrikeDelta(S,T,-0.10,rd,rf,vol10Put,inc_prem)
    
    return [[K10Put,K25Put,KATM,K25Call,K10Call], [vol10Put,vol25Put,ATM,vol25Call,vol10Call]]

def BuildSpline(Ks,vols):
    [K10Put,K25Put,KATM,K25Call,K10Call]=Ks
    [vol10Put,vol25Put,volATM,vol25Call,vol10Call]=vols

    a00 = 2.0/(K25Put-K10Put)
    a01 = 1.0/(K25Put-K10Put)
    a10 = 1.0/(K25Put-K10Put)
    a11 = 2.0/(K25Put-K10Put)+2.0/(KATM-K25Put)
    a12 = 1.0/(KATM-K25Put)
    a21 = 1.0/(KATM-K25Put)
    a22 = 2.0/(K25Call-KATM)+2.0/(KATM-K25Put)
    a23 = 1.0/(K25Call-KATM)
    a32 = 1.0/(K25Call-KATM)
    a33 = 2.0/(K25Call-KATM)+2.0/(K10Call-K25Call)
    a34 = 1.0/(K10Call-K25Call)
    a43 = 1.0/(K10Call-K25Call)
    a44 = 2.0/(K10Call-K25Call)
    
    b0 = 3.0*(vol25Put-vol10Put)/(K25Put-K10Put)/(K25Put-K10Put)
    b1 = 3.0*(vol25Put-vol10Put)/(K25Put-K10Put)/(K25Put-K10Put)+3.0*(volATM-vol25Put)/(KATM-K25Put)/(KATM-K25Put)
    b2 = 3.0*(volATM-vol25Put)/(KATM-K25Put)/(KATM-K25Put)+3.0*(vol25Call-volATM)/(K25Call-KATM)/(K25Call-KATM)
    b3 = 3.0*(vol25Call-volATM)/(K25Call-KATM)/(K25Call-KATM)+3.0*(vol10Call-vol25Call)/(K10Call-K25Call)/(K10Call-K25Call)
    b4 = 3.0*(vol10Call-vol25Call)/(K10Call-K25Call)/(K10Call-K25Call)
    
    A = np.array([[a00,a01,0,0,0],[a10,a11,a12,0,0],[0,a21,a22,a23,0],[0,0,a32,a33,a34],[0,0,0,a43,a44]])
    B =  np.array([b0,b1,b2,b3,b4])
    
    (k0,k1,k2,k3,k4) = k = tuple(np.linalg.solve(A,B))
    
    a = [0,0,0,0]
    b = [0,0,0,0]
    
    a[0]  = k0*(K25Put-K10Put)-(vol25Put-vol10Put)
    b[0]  = -k1*(K25Put-K10Put)+(vol25Put-vol10Put)
    a[1] = k1*(KATM-K25Put)-(volATM-vol25Put)
    b[1] = -k2*(KATM-K25Put)+(volATM-vol25Put)
    a[2] = k2*(K25Call-KATM)-(vol25Call-volATM)
    b[2] = -k3*(K25Call-KATM)+(vol25Call-volATM)
    a[3] = k3*(K10Call-K25Call)-(vol10Call-vol25Call)
    b[3] = -k4*(K10Call-K25Call)+(vol10Call-vol25Call)
    return (a,b,k)

def spline(x,x1,x2,y1,y2,a,b):
    t = (x-x1)/(x2-x1)
    return y1*(1.0-t)+y2*t+t*(1.0-t)*(a*(1.0-t)+b*t)
```

```python
def xfunc(x, pos=0):
         return '%1.2f'%(x)
        
def yfunc(y, pos=0):
         return '%1.0f'%(y*100)+'\%'


class SmileFXModel:
    def __init__(self,S,T,rd,rf,ATM,RR25,RR10,BF25,BF10,inc_prem=False):
        self.S = S
        self.T = T
        self.rd = rd
        self.rf = rf
        (self.Ks,self.vols) = SmileFX(S,T,rd,rf,ATM,RR25,RR10,BF25,BF10,inc_prem)
        (self.a,self.b,self.k) = BuildSpline(self.Ks,self.vols)
        self.inc_prem = inc_prem
        
    def vol(self,k):
        
        if k < self.Ks[0]:
            return spline(k,self.Ks[0],self.Ks[1],self.vols[0],self.vols[1],0.0,0.0)
        if k < self.Ks[1]:
            return spline(k,self.Ks[0],self.Ks[1],self.vols[0],self.vols[1],self.a[0],self.b[0])
        elif k < self.Ks[2]:
            return spline(k,self.Ks[1],self.Ks[2],self.vols[1],self.vols[2],self.a[1],self.b[1])
        elif k < self.Ks[3]:
            return spline(k,self.Ks[2],self.Ks[3],self.vols[2],self.vols[3],self.a[2],self.b[2])
        elif k < self.Ks[4]:
            return spline(k,self.Ks[3],self.Ks[4],self.vols[3],self.vols[4],self.a[3],self.b[3])
        else:
            return spline(k,self.Ks[3],self.Ks[4],self.vols[3],self.vols[4],0.0,0.0)
        
    def Option(self,K,cp,ccy):
        if ccy == 1:
            return BSFX(self.S,K,self.T,cp,self.rd,self.rf,self.vol(K))/S
        elif ccy == 2:
            return BSFX(self.S,K,self.T,cp,self.rd,self.rf,self.vol(K))
        
    def Digital(self,K,cp,ccy):
        eps = 1e-8
        if cp == 1:
            return (self.Option(K,cp,ccy)-self.Option(K+eps,cp,ccy))/eps
        else:
            return (self.Option(K+eps,cp,ccy)-self.Option(K,cp,ccy))/eps
        
    def plot_smile(self,ax,NbStdDev=2,color='black',linestyle='-'):
        FATM=self.Ks[2]
        volATM = self.vols[2]
        ax.xaxis.set_major_formatter(FuncFormatter(xfunc))
        ax.yaxis.set_major_formatter(FuncFormatter(yfunc))
        KMin=FATM*exp(-NbStdDev*volATM*sqrt(T))
        KMax=FATM*exp(+NbStdDev*volATM*sqrt(T))
        x = np.linspace(KMin, KMax, 40)
        y=[self.vol(k) for k in x]
        ax.plot(x,y,color=color,linestyle=linestyle)
        ax.scatter(self.Ks,self.vols,s=150, facecolors='green', edgecolors='green')
        
        ax.annotate(u'$(K^{10Put},\sigma^{10Put})$', xy=(self.Ks[0], self.vols[0]),  xycoords='data',
                    xytext=(10, 0), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='bottom',
                    )
        
        ax.annotate(u'$(K^{25Put},\sigma^{25Put})$', xy=(self.Ks[1], self.vols[1]),  xycoords='data',
                    xytext=(10, 0), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='bottom',
                    )
        
        ax.annotate(u'$(K^{ATM},\sigma^{ATM})$', xy=(self.Ks[2], self.vols[2]),  xycoords='data',
                    xytext=(-5, 0), textcoords='offset points',size=20,
                    horizontalalignment='right', verticalalignment='top',
                    )
        
        ax.annotate(u'$(K^{25Call},\sigma^{25Call})$', xy=(self.Ks[3], self.vols[3]),  xycoords='data',
                    xytext=(15, 5), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='top',
                    )
        
        ax.annotate(u'$(K^{10Call},\sigma^{10Call})$', xy=(self.Ks[4], self.vols[4]),  xycoords='data',
                    xytext=(15, 5), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='top',
                    )
        
        ax.set_title(u'Volatilité')
        
    def plot_option(self,ax,NbStdDev=2):
        FATM=self.Ks[2]
        volATM = self.vols[2]
        KMin=FATM*exp(-NbStdDev*volATM*sqrt(T))
        KMax=FATM*exp(+NbStdDev*volATM*sqrt(T))
        x = np.linspace(KMin, KMax, 40)
        y_call=[self.Option(k,1,2) for k in x]
        y_put=[self.Option(k,-1,2) for k in x]
        ax.plot(x,y_call,color='blue',label='call')
        ax.plot(x,y_put,color='red',label='put')
        ax.set_title(u'Call \& Put')
        ax.legend(loc='lower right')
        
    def plot_digital(self,ax,NbStdDev=2):
        FATM=self.Ks[2]
        volATM = self.vols[2]
        KMin=FATM*exp(-NbStdDev*volATM*sqrt(T))
        KMax=FATM*exp(+NbStdDev*volATM*sqrt(T))
        x = np.linspace(KMin, KMax, 40)
        y_call=[self.Digital(k,1,2) for k in x]
        y_put=[self.Digital(k,-1,2) for k in x]
        ax.plot(x,y_call,color='blue',label='digital call')
        ax.plot(x,y_put,color='red',label='digital put')
        ax.set_title(u'Digital Call \& Digital Put')
        ax.legend(loc='lower right')
    
    def ATM(self):
        return self.vols[2]
    def RR25(self):
        return self.vols[3]-self.vols[1]
    def RR10(self):
        return self.vols[4]-self.vols[0]
    def BF25(self):
        return (self.vols[3]+self.vols[1])*0.5-self.vols[2]
    def BF10(self):
        return (self.vols[4]+self.vols[0])*0.5-self.vols[2]
    
    def BumpSpot(self,bump=0.0001):
        return SmileFXModel(self.S+bump,
                            self.T,
                            self.rd,
                            self.rf,
                            self.ATM(),
                            self.RR25(),
                            self.RR10(),
                            self.BF25(),
                            self.BF10(),
                            self.inc_prem)
    
    def Bump_rd(self,bump=0.01):
        return SmileFXModel(self.S,
                            self.T,
                            self.rd+bump,
                            self.rf,
                            self.ATM(),
                            self.RR25(),
                            self.RR10(),
                            self.BF25(),
                            self.BF10(),
                            self.inc_prem)
    
    def Bump_rf(self,bump=0.01):
        return SmileFXModel(self.S,
                            self.T,
                            self.rd,
                            self.rf+bump,
                            self.ATM(),
                            self.RR25(),
                            self.RR10(),
                            self.BF25(),
                            self.BF10(),
                            self.inc_prem)
    
    def BumpATM(self,bump=0.01):
        return SmileFXModel(self.S,
                            self.T,
                            self.rd,
                            self.rf,
                            self.ATM()+bump,
                            self.RR25(),
                            self.RR10(),
                            self.BF25(),
                            self.BF10(),
                            self.inc_prem)
    
    def BumpRR(self,bump=0.01):
        if self.RR25() == 0:
            coeff = 2
        else:
            coeff = self.RR10()/self.RR25()
        return SmileFXModel(self.S,
                            self.T,
                            self.rd,
                            self.rf,
                            self.ATM(),
                            self.RR25()+bump,
                            self.RR10()+bump*coeff,
                            self.BF25(),
                            self.BF10(),
                            self.inc_prem)
    
    def BumpBF(self,bump=0.01):
        if self.BF25() == 0:
            coeff = 4
        else:
            coeff = self.BF10()/self.BF25()
        return SmileFXModel(self.S,
                            self.T,
                            self.rd,
                            self.rf,
                            self.ATM(),
                            self.RR25(),
                            self.RR10(),
                            self.BF25()+bump,
                            self.BF10()+bump*coeff,
                            self.inc_prem)
```

```python
def FXFwdRange(K,KMin,KMax,FXMod):
    price = 0
    price += (KMin-K)*FXMod.Digital(KMin,1,1)
    price += FXMod.Option(KMin,1,1)
    price -= FXMod.Option(KMax,1,1)
    price -= (KMax-K)*FXMod.Digital(KMax,1,1)
    return price

def ImpliedFXFwdRange(KMin,KMax,FXMod):
    eps = 1e-8
    K = FXMod.S
    for idx in range(20):
        val = FXFwdRange(K,KMin,KMax,FXMod)
        if abs(val)<eps:
            break
        der = (FXFwdRange(K+eps,KMin,KMax,FXMod)-val)/eps
        K = K - val/der
    
    return K
```
