```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
rcParams['text.latex.unicode']=True
```

```python
x1 = 1.3
y1 = 0.12
x2 = 1.5
y2 = 0.14
```

```python
def xfunc(x, pos=0):
         return '%1.2f'%(x)
        
def yfunc(y, pos=0):
         return '%1.1f'%(y*100)+'\%'


    
def figure_interpolation(filename,slide):
    fig,ax = plt.subplots(figsize=(10,5))
    ax.scatter(x1,y1,s=150, facecolors='green', edgecolors='green')
    ax.scatter(x2,y2,s=150, facecolors='green', edgecolors='green')
    
    ax.xaxis.set_major_formatter(FuncFormatter(xfunc))
    ax.yaxis.set_major_formatter(FuncFormatter(yfunc))
    
    ax.set_xlim([0.97*x1,1.03*x2])
    ax.set_ylim([0.97*y1,1.03*y2])
    
    x = np.linspace(x1*0.99, x2*1.01, 40)
    t = [(x_val-x1)/(x2-x1) for x_val in x]
    y_interp_lin = [ (1.0-t_val)*y1 + t_val*y2 for t_val in t]
    ax.plot(x,y_interp_lin,linestyle='--',color='black',label='linear')
    
    ax.set_xlabel('K',fontsize=17)
    ax.set_ylabel('$\sigma$',fontsize=17)
    
    ax.annotate(u'($x_1,y_1$)', xy=(x1, y1),  xycoords='data',
                    xytext=(-30, 30), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='bottom',
                    )
    
    ax.annotate(u'($x_2,y_2$)', xy=(x2, y2),  xycoords='data',
                    xytext=(0, -30), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='bottom',
                    )
    
    a_list = [-0.03,0.03]
    b_list = [-0.03,0.03]
    if slide == 1:
        b = 0
        for a in a_list:
            y_interp_cubic = [ (1.0-t_val)*y1 + t_val*y2 + t_val*(1-t_val)*(a*(1.0-t_val)+b*t_val) for t_val in t]
            ax.plot(x,y_interp_cubic,linestyle='--',color='black',linewidth=2,label='cubic, a=%1.0f'%(a*100)+'\\%')
    if slide == 2:
        a=0
        for b in b_list:
            y_interp_cubic = [ (1.0-t_val)*y1 + t_val*y2 + t_val*(1-t_val)*(a*(1.0-t_val)+b*t_val) for t_val in t]
            ax.plot(x,y_interp_cubic,linestyle='--',color='black',linewidth=2,label='cubic, b=%1.0f'%(b*100)+'\\%')
    
    legend = ax.legend(loc='lower right')
    for label in legend.get_texts():
        label.set_fontsize(17)
        
    plt.savefig(filename, bbox_inches='tight')
```

```python
figure_interpolation('../figures/interpolation-0.png',0)
```

```python
xmin = 1.1
ymin = 0.13
xmid = 1.3
ymid = 0.12
xmax = 1.5
ymax = 0.14

a00 = 2.0/(xmid-xmin)
a01 = 1.0/(xmid-xmin)
a10 = 1.0/(xmid-xmin)
a11 = 2.0/(xmid-xmin)+2.0/(xmax-xmid)
a12 = 1.0/(xmax-xmid)
a21 = 1.0/(xmax-xmid)
a22 = 2.0/(xmax-xmid)
b0=3.0*(ymid-ymin)/(xmid-xmin)/(xmid-xmin)
b1=3.0*(ymid-ymin)/(xmid-xmin)/(xmid-xmin)+3.0*(ymax-ymid)/(xmax-xmid)/(xmax-xmid)
b2=3.0*(ymax-ymid)/(xmax-xmid)/(xmax-xmid)

a = np.array([[a00, a01, 0], [a10, a11, a12], [0, a21, a22]])
b = np.array([b0, b1, b2])
```

```python
(k0, k1, k2)=tuple(np.linalg.solve(a,b))
a1  = k0*(xmid-xmin)-(ymid-ymin)
b1  = -k1*(xmid-xmin)+(ymid-ymin)
a2  = k1*(xmax-xmid)-(ymax-ymid)
b2  = -k2*(xmax-xmid)+(ymax-ymid)

def spline(t,y1,y2,a,b):
    return (1-t)*y1+t*y2+t*(1.0-t)*(a*(1.0-t)+b*t)
```

```python
def figure_interpolation2(filename,slide):
    fig,ax = plt.subplots(figsize=(10,5))
    ax.scatter(xmin,ymin,s=150, facecolors='green', edgecolors='green')
    ax.scatter(xmid,ymid,s=150, facecolors='green', edgecolors='green')
    ax.scatter(xmax,ymax,s=150, facecolors='green', edgecolors='green')
    
    ax.xaxis.set_major_formatter(FuncFormatter(xfunc))
    ax.yaxis.set_major_formatter(FuncFormatter(yfunc))
    
    ax.set_xlim([0.95*xmin,1.05*xmax])
    ax.set_ylim([0.95*ymid,1.05*ymax])
    
    x_min_mid = np.linspace(0.97*xmin, 1.03*xmid, 40)
    x_mid_max = np.linspace(0.97*xmid, 1.03*xmax, 40)
    t_min_mid = [(x_val-xmin)/(xmid-xmin) for x_val in x_min_mid]
    t_mid_max = [(x_val-xmid)/(xmax-xmid) for x_val in x_mid_max]
    y_min_mid_interp_lin = [ spline(t_val,ymin,ymid,a1,b1) for t_val in t_min_mid]
    y_mid_max_interp_lin = [ spline(t_val,ymid,ymax,a2,b2) for t_val in t_mid_max]
    ax.plot(x_min_mid,y_min_mid_interp_lin,linestyle='--',color='red',label='linear')
    ax.plot(x_mid_max,y_mid_max_interp_lin,linestyle='--',color='blue',label='linear')
    
    ax.set_xlabel('K',fontsize=17)
    ax.set_ylabel('$\sigma$',fontsize=17)
    
    ax.annotate(u'($x_{i-1},y_{i-1}$)', xy=(xmin, ymin),  xycoords='data',
                    xytext=(-30, 30), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='bottom',
                    )
    
    ax.annotate(u'($x_{i},y_{i}$)', xy=(xmid, ymid),  xycoords='data',
                    xytext=(-30, 30), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='bottom',
                    )
    
    ax.annotate(u'($x_{i+1},y_{i+1}$)', xy=(xmax, ymax),  xycoords='data',
                    xytext=(0, -30), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='bottom',
                    )
    
    
    ax.annotate(u'$q^i(x,a_i,b_i)$', xy=((xmin+xmid)/2, ymid),  xycoords='data',
                    color='red',
                    xytext=(-50, -20), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='bottom',
                    )
    
    ax.annotate(u'$q^{i+1}(x,a_{i+1},b_{i+1})$', xy=((xmid+xmax)/2, (ymid+ymax)/2),  xycoords='data',
                    color='blue',
                    xytext=(40, -20), textcoords='offset points',size=20,
                    horizontalalignment='left', verticalalignment='bottom',
                    )
        
    plt.savefig(filename, bbox_inches='tight')
```

```python
figure_interpolation2('../figures/interpolation-3.png',0)
```
