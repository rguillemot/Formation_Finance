```python
import matplotlib as mpl
```

```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
mpl.rcParams['text.latex.unicode']=True
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plotFixFlow(ax,starty,t,value,color,label):
    ax.arrow(t,starty,0,value,color=color,head_width=0.02, head_length=0.05,linewidth=0.2)
    ax.annotate(label, xy=(t,starty),  
                xycoords='data', size=20, xytext=(10, -25), textcoords='offset points'
                )    
def plotdelta(ax,starty,t1,t2,color,label):
    x = np.arange(t1,t2+0.02,0.02)
    y = starty-(x-t1)*(x-t2)/(t1-t2)/(t1-t2)*0.3
    ax.plot(x,y,color=color,linewidth=0.4,linestyle='--')
    ax.annotate(label, xy=((t1+t2)/2,starty),  
                xycoords='data', size=25, xytext=(-50, 50), textcoords='offset points'
                )

def figure(slide,filename):
    fig, ax = plt.subplots(figsize=(20,10))
    
    ax.set_xlim([-0.1,0.7])
    ax.set_ylim([-0.5,0.5])
    
    plt.axis('off')
    
    ax.arrow(0,-0.02,0,0.04,color='black',head_width=0, head_length=0,linewidth=0.1)
    ax.annotate('$t$', xy=(0,0),  
                    xycoords='data', size=20, xytext=(-20, -35), textcoords='offset points'
                    )
    
    ax.arrow(0.5,-0.02,0,0.04,color='black',head_width=0, head_length=0,linewidth=0.05)
    ax.annotate('$t$+2D', xy=(0.5,0),  
                    xycoords='data', size=20, xytext=(-50, -35), textcoords='offset points',
                    )
    
    ax.arrow(-0.5,0,1.1,0,color='black',head_width=0.05, head_length=0.02,linewidth=0.05)
    
    
    if slide>=3:
        plotdelta(ax,0,0,0.5,u'black',u'2 jours ouvrés.')
    
    if slide>=4:
        plotFixFlow(ax,0,0.5,0.2,'red','')
        plotFixFlow(ax,0,0.5,-0.2,'blue','')
    
    if slide>=2:
        ax.annotate(u'Date d\'exécution \n Trade Date', xy=(0, 0.05),  xycoords='data',
                        xytext=(0.15, 0.8), textcoords='axes fraction',
                        size=25,
                        fontweight='extra bold',
                        arrowprops=dict(arrowstyle="->",facecolor='black'),
                        horizontalalignment='left', verticalalignment='top',
                        )
    
    if slide>=3:
        ax.annotate(u'Date de livraison \n Settlement Date', xy=(0.47, 0.05),  xycoords='data',
                        xytext=(0.55, 0.8), textcoords='axes fraction',
                        size=25,
                        arrowprops=dict(arrowstyle="->",facecolor='black'),
                        horizontalalignment='left', verticalalignment='top',
                        )
    
    if slide >= 4:
        ax.annotate('1.1162 dollar', xy=(0.5,0), color = 'blue',  
                        xycoords='data', size=20, xytext=(+20, -50), textcoords='offset points'
                        )
        
        
        ax.annotate('1 euro', xy=(0.5,0), color = 'red',  
                        xycoords='data', size=20, xytext=(+20, +50), textcoords='offset points'
                        )
    
    plt.savefig(filename, bbox_inches='tight')
```
