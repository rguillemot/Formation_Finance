```python
T=1.0
S=1.3889
rEUR=0.005
rUSD=0.003
m = 0.001
vol = 0.12

import numpy as np
from scipy.stats import norm
import math

def BSFX(S,K,T,cp,rd,rf,vol):
    if vol*T == 0:
        return cp*(F-K)
    else:
        F = S*exp((rd-rf)*T)
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        d2 = d1 - vol*math.sqrt(T)
    return (cp*F*norm.cdf(cp*d1)-cp*K*norm.cdf(cp*d2))*exp(-rd*(T))

def Delta(S,K,T,cp,rd,rf,vol):
    F = S*exp((rd-rf)*T)
    d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
    return cp*norm.cdf(cp*d1)*exp(-rf*(T))

def Vega(S,K,T,cp,rd,rf,vol):
    F = S*exp((rd-rf)*T)
    d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
    return S*sqrt(T)*norm.pdf(cp*d1)*exp(-rf*(T))    
```

```python
fig, ((ax1,ax2,ax3)) = plt.subplots(1,3,figsize=(12,4))

def xfunc(x, pos=0):
         return '%1.2f'%(x)
        
def yfunc(y, pos=0):
         return '%1.0f'%(y*100)+'\%'
    
def yfunc_no100(y, pos=0):
         return '%1.2f'%(y)+'\%'

Ks = np.linspace(0.9,2.0,100)
Deltas = [Delta(S,K,T,1,rUSD,rEUR-m,vol) for K in Ks]
Vegas = [Vega(S,K,T,1,rUSD,rEUR-m,vol) for K in Ks]

ax1.xaxis.set_major_formatter(FuncFormatter(xfunc))
ax1.yaxis.set_major_formatter(FuncFormatter(yfunc))

ax2.xaxis.set_major_formatter(FuncFormatter(xfunc))
ax2.yaxis.set_major_formatter(FuncFormatter(yfunc_no100))

ax3.xaxis.set_major_formatter(FuncFormatter(yfunc))
ax3.yaxis.set_major_formatter(FuncFormatter(yfunc_no100))

ax1.set_title('Delta',fontsize=17)
ax2.set_title('Vega',fontsize=17)
ax3.set_title('Vega=f($\delta$)',fontsize=17)
ax1.set_xlabel('Strike',fontsize=17)
ax2.set_xlabel('Strike',fontsize=17)
ax3.set_xlabel('Delta',fontsize=17)

ax1.set_xlim(Ks[0],Ks[-1])
ax2.set_xlim(Ks[0],Ks[-1])
ax3.set_xlim(Deltas[0],Deltas[-1])

ax1.plot(Ks,Deltas,color='black')
ax2.plot(Ks,Vegas,color='black')
ax3.plot(Deltas,Vegas,color='black')
plt.savefig('../figures/fxopt-deltavsvega', bbox_inches='tight')
```
