```python
from matplotlib import rc, rcParams
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
rcParams['text.latex.unicode']=True
```

```python
NEUR = 100000000
NUSD = 109000000

KMin = 0.95
KMax = 1.25
PMin = -16000000
PMax = 16000000

def xfunc(x, pos=0):
         return '%1.2f'%(x)
def yfunc(y, pos=0):
     return '%1.0f Mios'%(y/1e6)


def format_graph(ax,x,y,KMin,KMax,PMin,PMax,title,xtitle,ytitle,color):
    ax.set_xlim([KMin,KMax])
    ax.set_ylim([PMin,PMax])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    ax.spines['bottom'].set_position('zero')
    ax.xaxis.set_major_formatter(FuncFormatter(xfunc))
    ax.yaxis.set_major_formatter(FuncFormatter(yfunc))
    ax.set_title(title)
    ax.set_xlabel(xtitle)
    ax.set_ylabel(ytitle)
    ax.plot(x,y,color=color,linewidth=3)


def payoff(filename,slide, cp):
    fig, ax = plt.subplots(figsize=(9,7))
    
    K=float(NUSD)/NEUR
    EURUSD = np.linspace(KMin, KMax, 100)
    EURUSDCall = np.linspace(K, KMax, 100)
    EURUSDPut = np.linspace(KMin, K, 100)
    EURUSDPNL=[(NEUR*s-NUSD)*cp for s in EURUSD]
    EURUSDCallPNL=[(NEUR*s-NUSD)*cp for s in EURUSDCall]
    EURUSDPutPNL=[(NEUR*s-NUSD)*cp for s in EURUSDPut]
    
    if slide >= 2:
        format_graph(ax,EURUSD,EURUSDPNL,KMin,KMax,PMin,PMax,'Payoff','EUR/USD','USD','black')
        verts = [(KMin, 0)] + list(zip(EURUSDPut, EURUSDPutPNL)) + [(K, 0)]
        poly = Polygon(verts, facecolor='red', edgecolor='0.5',alpha=0.5, hatch='/')
        ax.add_patch(poly)
    elif slide >= 1:
        format_graph(ax,EURUSD,EURUSDPNL,KMin,KMax,PMin,PMax,'Payoff','EUR/USD','USD','black')
        verts = [(K, 0)] + list(zip(EURUSDCall, EURUSDCallPNL)) + [(KMax, 0)]
        poly = Polygon(verts, facecolor='red', edgecolor='0.5',alpha=0.5, hatch='/')
        ax.add_patch(poly)
    else:
        format_graph(ax,EURUSD,EURUSDPNL,KMin,KMax,PMin,PMax,'EUR/USD - USD','EUR/USD','USD','black')
        
    plt.savefig(filename, bbox_inches='tight')
   
```
