```python
import numpy as np
from scipy.stats import norm
import math

def BSFX(S,K,T,cp,rd,rf,vol):
    DFd = exp(-rd*T)
    DFf = exp(-rf*T)
    F = S*DFf/DFd
    if vol*T == 0:
        return max(cp*DFd*(F-K),0)
    else:
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        d2 = d1 - vol*math.sqrt(T)
        return (cp*F*norm.cdf(cp*d1)-cp*K*norm.cdf(cp*d2))*DFd

def BSDelta(S,K,T,cp,rd,rf,vol):
    DFd = exp(-rd*T)
    DFf = exp(-rf*T)
    if vol*T == 0:
        if cp*S > cp*K:
            return cp*DFf
        else:
            return 0.0
    else:
        F = S*DFf/DFd
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        return cp*norm.cdf(cp*d1)*DFf
```

```python
NEUR = 100000000.0
NUSDATM = 139750000.0
NUSDCall = 151478455.0
NUSDPut = 127548288.0

S = 1.3889
rUSD = 0.003
rEUR =0.005
m=0.001
vol = 0.12
KCall = float(NUSDCall)/NEUR
KPut = float(NUSDPut)/NEUR
KATM = float(NUSDATM)/NEUR
T = 1.0

KMin = 1.2
KMax = 1.6
PMin = -1000000
PMax = 15000000

def xfunc(x, pos=0):
         return '%1.2f'%(x)
    
def yfunc(y, pos=0):
     return '%1.0f Mios'%(y/1e6)
    
def yfunc_delta(y, pos=0):
         return '%1.0f'%(y*100)+'\%'


def format_graph(ax,x,y,KMin,KMax,PMin,PMax,title,xtitle,ytitle,color,delta=False):
    ax.set_xlim([0.99*KMin,1.01*KMax])
    ax.set_ylim([PMin,PMax])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    ax.spines['bottom'].set_position('zero')
    ax.xaxis.set_major_formatter(FuncFormatter(xfunc))
    if delta:
        ax.yaxis.set_major_formatter(FuncFormatter(yfunc_delta))
    else:
        ax.yaxis.set_major_formatter(FuncFormatter(yfunc))
    ax.set_title(title,fontsize=20)
    ax.set_xlabel(xtitle)
    ax.set_ylabel(ytitle)
    ax.plot(x,y,color=color,linewidth=3)
    
def delta_EUR_USD(K):
    delta = BSDelta(S,K,T,1,rUSD,rEUR-m,vol)+BSDelta(S,K,T,-1,rUSD,rEUR-m,vol)
    return delta
    
def butterfly(filename,slide):
    fig,ax = plt.subplots(figsize=(10,5))
    
    EURUSD = np.linspace(KMin, KMax, 100)
    payoff=[max(NEUR*s-NUSDCall,0)+max(NEUR*s-NUSDPut,0)-2.0*max(NEUR*s-NUSDATM,0) for s in EURUSD]
    value=[1*BSFX(s,KCall,T,1,rUSD,rEUR-m,vol)*NEUR+1*BSFX(s,KPut,T,1,rUSD,rEUR-m,vol)*NEUR-2*BSFX(s,KATM,T,1,rUSD,rEUR-m,vol)*NEUR for s in EURUSD]
    
    if slide>=1:
        format_graph(ax,EURUSD,payoff,KMin,KMax,PMin,PMax,u'Payoff et valeur','F (EUR/USD)','USD','black')
        ax.annotate(u'$K^{25Call}$', xy=(KCall+0.01, +600000),  xycoords='data',
                    xytext=(+20, +40), textcoords='offset points',size=20,
                    arrowprops=dict(arrowstyle="->",facecolor='black'),
                    horizontalalignment='left', verticalalignment='bottom',
                    )
        ax.annotate(u'$K^{25Put}$', xy=(KPut-0.01, +300000),  xycoords='data',
                    xytext=(-20, +40), textcoords='offset points',size=20,
                    arrowprops=dict(arrowstyle="->",facecolor='black'),
                    horizontalalignment='right', verticalalignment='bottom',
                    )
        
        ax.annotate(u'$K^{ATM}$', xy=(KATM, ((KATM-KPut))*NEUR-300000),  xycoords='data',
                    xytext=(-25, -50), textcoords='offset points',size=20,
                    arrowprops=dict(arrowstyle="->",facecolor='black'),
                    horizontalalignment='middle', verticalalignment='bottom',
                    )
    else:
        format_graph(ax,[],[],KMin,KMax,PMin,PMax,u'Payoff et valeur','F (EUR/USD)','USD','black')
    if slide>=2:
        format_graph(ax,EURUSD,value,KMin,KMax,PMin,PMax,u'Payoff et valeur','F (EUR/USD)','USD','blue')
    else:
        format_graph(ax,[],[],KMin,KMax,PMin,PMax,u'Payoff et valeur','F (EUR/USD)','USD','blue')
        
        
    plt.savefig(filename, bbox_inches='tight')
```
