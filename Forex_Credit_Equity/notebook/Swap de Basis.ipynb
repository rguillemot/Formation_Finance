```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
rcParams['text.latex.unicode']=True
```

```python
import numpy as np
import matplotlib.pyplot as plt


def plotFixFlow(ax,startx,t,value,color,label,size,posx=-1):
    ax.arrow(startx,t,value,0,color=color,head_width=0.04, head_length=0.02)
    if value<0:
        posx = 20
    else:
        if posx == - 1:
            posx = -340
            
    ax.annotate(
                label, 
                xy=(startx,t),
                xycoords='data',
                color = color, 
                size=size, 
                xytext=(posx, -10), 
                textcoords='offset points'
    )
    
def plotVarFlow(ax,startx,t,value,color,label,size):
    x = startx + np.arange(0,abs(value)+0.001,0.001)*value/abs(value)
    y = [t+np.sin(v/value*np.pi*8)*0.01 for v in x]
    ax.plot(x,y,color=color)
    if value > 0:
        dec = 0.002
    else:
        dec = -0.002
    ax.arrow(startx+value+dec,t,dec/abs(dec)*0.0001,0,color=color,head_width=0.04, head_length=0.02,linewidth=2.5)
    if value<0:
        posx = 20
    else:
        posx = -340
    ax.annotate(
                label, 
                xy=(startx,t),
                xycoords='data',
                color = color, 
                size=size, 
                xytext=(posx, -10), 
                textcoords='offset points'
    )
    
def plotdelta(ax,startx,t1,t2,color,label):
    y = np.arange(t1,t2+0.02,0.02)
    x = startx+(y-t1)*(y-t2)/(t1-t2)/(t1-t2)*0.05
    ax.plot(x,y,color=color,linewidth=0.4,linestyle='--')
    ax.annotate(label, xy=(startx,(t1+t2)/2),  
                xycoords='data', size=25, xytext=(-50, -5), textcoords='offset points'
                )
def plotdash(ax,x,t1,t2,color):
    y = np.arange(t1,t2+0.02,0.02)
    x = [x for i in range(len(y))]
    ax.plot(x,y,color=color,linewidth=0.4,linestyle='--')


    
def figure(slide,filename):
    fig, ax = plt.subplots(figsize=(20,10))
    
    ax.set_xlim([0,1])
    ax.set_ylim([-0.5,0.5])
    
    plt.axis('off')

    xEUR = 0.87
    xUSD = 0.12

    ax.arrow(xUSD,-0.4,0,0.8,color='black',head_width=0, head_length=0)
    ax.arrow(xEUR,-0.4,0,0.8,color='black',head_width=0, head_length=0)
    
    ax.arrow(0.5,-0.4,0,0.8,color='black',head_width=0.02, head_length=0.04)
    
    ax.arrow(0.49,0.35,0.02,0,color='black',head_width=0, head_length=0,linewidth=0.05)
    
    ax.annotate('$T_N$', xy=(0.5,0.35),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    ax.arrow(0.49,-0.35,0.02,0,color='black',head_width=0, head_length=0)
    
    ax.annotate('$T_0=$t+2D', xy=(0.5,-0.35),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    ax.arrow(0.49,-0.37,0.02,0,color='black',head_width=0, head_length=0)
    
    ax.annotate('t', xy=(0.5,-0.40),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    ax.arrow(0.49,-0.25,0.02,0,color='black',head_width=0, head_length=0)
    
    ax.annotate('$T_1$', xy=(0.5,-0.25),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    ax.arrow(0.49,-0.15,0.02,0,color='black',head_width=0, head_length=0)
    
    ax.annotate('$T_2$', xy=(0.5,-0.15),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    ax.arrow(0.49,0.25,0.02,0,color='black',head_width=0, head_length=0)
    
    ax.annotate('$T_{n-1}$', xy=(0.5,0.25),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    
    ax.annotate('USD', xy=(xUSD,0.45),  
                    xycoords='data', size=25, xytext=(-20, 0), textcoords='offset points'
                    )
    
    ax.annotate('EUR', xy=(xEUR,0.45),  
                    xycoords='data', size=25, xytext=(-20, 0), textcoords='offset points'
                    )
    
    ax.annotate('Temps', xy=(0.5,0.45),  
                    xycoords='data', size=25, xytext=(-40, 5), textcoords='offset points'
                    )
    
    if slide >= 2:
        plotFixFlow(ax,xUSD,-0.35,0.1,'blue','$N^{USD}$',30,-100)
        plotFixFlow(ax,xEUR,-0.35,-0.1,'red','$N^{EUR}$',30)
    if slide == 2:
        plotFixFlow(ax,xUSD,0.35,-0.1,'blue','$N^{USD}$',30)
        plotFixFlow(ax,xEUR,0.35,0.1,'red','$N^{EUR}$',30,-100)
    elif slide >=3:
        plotFixFlow(ax,xUSD,0.35,-0.1,'blue','$N^{USD}+N^{USD} \delta_N L^{USD}_N$',30)
        if slide == 3:
            plotFixFlow(ax,xEUR,0.35,0.1,'red','$N^{EUR} \delta_N L^{EUR}_N+N^{EUR}$',30)
        elif slide >= 4:
            plotFixFlow(ax,xEUR,0.35,0.1,'red','$N^{EUR} [ 1 + \delta_N (L^{EUR}_N-\\textbf{m})]$',30)
    
    if slide >=3:
        plotVarFlow(ax,xUSD,-0.25,-0.05,'blue','$N^{USD} \delta_1 L^{USD}_1$',30)
        plotVarFlow(ax,xUSD,-0.15,-0.05,'blue','$N^{USD} \delta_2 L^{USD}_2$',30)
        plotVarFlow(ax,xUSD,0.25,-0.05,'blue','$N^{USD} \delta_{N-1} L^{USD}_{N-1}$',30)
        plotVarFlow(ax,xUSD,0.32,-0.05,'blue','',30)
        if slide == 3:
            plotVarFlow(ax,xEUR,-0.25,0.05,'red','$N^{EUR} \delta_1 L^{EUR}_1$',30)
            plotVarFlow(ax,xEUR,-0.15,0.05,'red','$N^{EUR} \delta_2 L^{EUR}_2$',30)
            plotVarFlow(ax,xEUR,0.25,0.05,'red','$N^{EUR} \delta_{N-1} L^{EUR}_{N-1}$',30)
            plotVarFlow(ax,xEUR,0.32,0.05,'red','',30)
        elif slide >= 4:
            plotVarFlow(ax,xEUR,-0.25,0.05,'red','$N^{EUR} \delta_1 (L^{EUR}_1-\\textbf{m})$',30)
            plotVarFlow(ax,xEUR,-0.15,0.05,'red','$N^{EUR} \delta_2 (L^{EUR}_2-\\textbf{m})$',30)
            plotVarFlow(ax,xEUR,0.25,0.05,'red','$N^{EUR} \delta_{N-1} (L^{EUR}_{N-1}-\\textbf{m})$',30)
            plotVarFlow(ax,xEUR,0.32,0.05,'red','',30)
    
    plotdelta(ax,0.5,-0.35,-0.25,'black','$\delta_1$')
    plotdelta(ax,0.5,-0.25,-0.15,'black','$\delta_2$')
    plotdelta(ax,0.5,0.25,0.35,'black','$\delta_N$')
    
    plotdash(ax,0.55,-0.1,0.2,'black')
    if slide>=3:
        plotdash(ax,xUSD-0.05,-0.1,0.2,'blue')
        plotdash(ax,xEUR+0.05,-0.1,0.2,'red')
    
    plt.savefig(filename, bbox_inches='tight')
```

```python
def figureSingleFlow(filename):
    fig, ax = plt.subplots(figsize=(20,10))
    
    ax.set_xlim([0,1])
    ax.set_ylim([-0.5,0.5])
    
    plt.axis('off')

    xEUR = 0.87
    xUSD = 0.12

    ax.arrow(xUSD,-0.4,0,0.8,color='black',head_width=0, head_length=0)
    ax.arrow(xEUR,-0.4,0,0.8,color='black',head_width=0, head_length=0)
    
    ax.arrow(0.5,-0.4,0,0.8,color='black',head_width=0.02, head_length=0.04)
    
    ax.arrow(0.49,0.35,0.02,0,color='black',head_width=0, head_length=0,linewidth=0.05)
    
    ax.annotate('$T$', xy=(0.5,0.35),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    ax.arrow(0.49,-0.35,0.02,0,color='black',head_width=0, head_length=0)
    
    ax.annotate('$t+2D$', xy=(0.5,-0.35),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    ax.arrow(0.49,-0.37,0.02,0,color='black',head_width=0, head_length=0)
    
    ax.annotate('t', xy=(0.5,-0.40),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    ax.annotate('USD', xy=(xUSD,0.45),  
                    xycoords='data', size=25, xytext=(-20, 0), textcoords='offset points'
                    )
    
    ax.annotate('EUR', xy=(xEUR,0.45),  
                    xycoords='data', size=25, xytext=(-20, 0), textcoords='offset points'
                    )
    
    ax.annotate('Temps', xy=(0.5,0.45),  
                    xycoords='data', size=25, xytext=(-40, 5), textcoords='offset points'
                    )
    
    plotFixFlow(ax,xUSD,-0.35,0.1,'blue','$N^{USD}$',30,-100)
    plotFixFlow(ax,xEUR,-0.35,-0.1,'red','$N^{EUR}$',30)
    plotFixFlow(ax,xUSD,0.35,-0.1,'blue','$N^{USD}+N^{USD} \delta L^{USD}$',30)
    plotFixFlow(ax,xEUR,0.35,0.1,'red','$N^{EUR} [ 1 + \delta (L^{EUR}-\\textbf{m})]$',30)
    
    plotVarFlow(ax,xUSD,0.32,-0.05,'blue','',30)
    plotVarFlow(ax,xEUR,0.32,0.05,'red','',30)
    
        
    plotdelta(ax,0.5,-0.35,0.35,'black','$\delta$')
    
    
    plt.savefig(filename, bbox_inches='tight')
```
