```python
import numpy as np
import matplotlib.pyplot as plt

def plotFixFlow(ax,starty,t,value,color,label):
    ax.arrow(t,starty,0,value,color=color,head_width=0.05, head_length=0.1,linewidth=0.4)
    ax.annotate(label, xy=(t,starty),  
                xycoords='data', size=20, xytext=(10, 35), textcoords='offset points'
                )


def plotVarFlow(ax,starty,t,value,color,label):
    y = starty + np.arange(0,value+0.005,0.005)
    x = [t+np.sin(v/value*np.pi*10)*0.02 for v in y]
    ax.plot(x,y,color='black',linewidth=0.4)
    ax.arrow(t,starty+value+0.02,0,0.0001,color=color,head_width=0.05, head_length=0.1,linewidth=2.5)
    ax.annotate(label, xy=(t,starty),  
                xycoords='data', size=20, xytext=(10, -35), textcoords='offset points'
                )
    
def plotdelta(ax,starty,t1,t2,color,label,coeff=1):
    x = np.arange(t1,t2+0.02,0.02)
    y = starty-(x-t1)*(x-t2)/(t1-t2)/(t1-t2)*0.3*coeff
    ax.plot(x,y,color=color,linewidth=0.4,linestyle='--')
    ax.annotate(label, xy=((t1+t2)/2,starty),  
                xycoords='data', size=20, xytext=(-10, coeff*15), textcoords='offset points'
                )


fig, ax = plt.subplots(figsize=(15,3))

ax.set_xlim([-0.5,3])
ax.set_ylim([0.7,2])

plt.axis('off')

ax.arrow(0,0.9,0,0.2,color='black',linewidth=0.1)
ax.annotate('$T_0=0$', xy=(0,1),  
                xycoords='data', size=20, xytext=(-30, -25), textcoords='offset points'
                )

ax.arrow(2,0.9,0,0.2,color='black',linewidth=0.1)
ax.annotate('$T_n=T$', xy=(2,1),  
                xycoords='data', size=20, xytext=(+10, -25), textcoords='offset points'
                )

ax.arrow(-0.5,1,3,0,color='black',head_width=0.2, head_length=0.1,linewidth=0.1)

plotdelta(ax,1.3,1.12,1.9,'black','',0)

plotFixFlow(ax,1,0.3,0.4,'black','$T_1$')
plotFixFlow(ax,1,0.6,0.4,'black','$T_2$')
plotFixFlow(ax,1,0.9,0.4,'black','$T_3$')
plotFixFlow(ax,1,2,0.8,'black','$T_n$')

plotdelta(ax,1,0,0.3,'black','$\delta$')
plotdelta(ax,1,0.3,0.6,'black','$\delta$') 
plotdelta(ax,1,0.6,0.9,'black','$\delta$')

ax.annotate('$\\delta C$', xy=(0.4,1.2),  
                xycoords='data', size=17, xytext=(-30, 50), textcoords='offset points'
                )

ax.annotate('$\\delta C$', xy=(0.7,1.2),  
                xycoords='data', size=17, xytext=(-30, 50), textcoords='offset points'
                )

ax.annotate('$\\delta C$', xy=(1,1.2),  
                xycoords='data', size=17, xytext=(-30, 50), textcoords='offset points'
                )

ax.annotate('$1+\\delta C$', xy=(2,1.2),  
                xycoords='data', size=17, xytext=(10, 50), textcoords='offset points'
                )


plt.savefig('../figures/bond.png', bbox_inches='tight')
```

```python
def xfunc(x, pos=0):
         return '%1.0f'%(x*100)+'\%'
    
def yfunc(y, pos=0):
         return '%1.0f'%(y*100)


def bond_price(C,delta,n,R):
    return C/R*(1.0-1.0/((1.0+R/delta)**(n/delta)))+1.0/((1.0+R/delta)**(n/delta))

x = np.linspace(0.005,0.1,100)
y = [bond_price(0.05,1,10,R) for R in x]


fig, ax = plt.subplots(figsize=(8,8))

ax.set_xlim(0.001,0.11)
ax.set_ylim(0.50,1.50)

ax.xaxis.set_major_formatter(FuncFormatter(xfunc))
ax.yaxis.set_major_formatter(FuncFormatter(yfunc))

ax.plot(x,y,color='black',linewidth=2)
ax.set_title('Relation Prix versus Rendement.', fontsize=17)
ax.set_xlabel('Rendement (R)',fontsize=15)
ax.set_ylabel('Prix (P)',fontsize=15)

x1 = np.linspace(0.001,0.05,20)
y1 = [1] * 20
x2 = [0.05] * 20
y2 = np.linspace(0,1,20)

ax.plot(x1,y1,linestyle='--',color='black')
ax.plot(x2,y2,linestyle='--',color='black')
ax.plot([5],[100],marker='o',color='black',markersize=17)

ax.annotate('Obligation au pair $P(5\%)=100$', xy=(0.05,1),  
                xycoords='data', size=20, xytext=(-45, 40), textcoords='offset points'
                )
plt.savefig('../figures/prixvsrdt.png', bbox_inches='tight')
```

```python
def DirtyPrice(leg,t,C,R):
    ret = 0
    for sd,ed in zip(leg[:-1],leg[1:]):
        if t < ed:
                ret = ret + C/(1.0+(ed-sd)*R)**((ed-t)/(ed-sd))
    ret = ret + 1.0/(1.0+(ed-sd)*R)**((ed-t)/(ed-sd))
    return ret

def Accrual(leg,t,C,R):
    ret = 0
    for sd,ed in zip(leg[:-1],leg[1:]):
        if t>sd and t < ed:
                ret = C*(t-sd)/(ed-sd)
                break
    return ret

def CleanPrice(leg,t,C,R):
    return DirtyPrice(leg,t,C,R)-Accrual(leg,t,C,R)

leg = np.linspace(0,10,11)
times = np.linspace(0,10,101)
dprices_4 = [DirtyPrice(leg,t,0.05,0.04) for t in times]
dprices_5 = [DirtyPrice(leg,t,0.05,0.05) for t in times]
dprices_6 = [DirtyPrice(leg,t,0.05,0.06) for t in times]

cprices_4 = [CleanPrice(leg,t,0.05,0.04) for t in times]
cprices_5 = [CleanPrice(leg,t,0.05,0.05) for t in times]
cprices_6 = [CleanPrice(leg,t,0.05,0.06) for t in times]


fig, ((ax1,ax2,ax3)) = plt.subplots(1,3,figsize=(15,5),sharey=True)
ax1.plot(times,dprices_4,linestyle='--',color='black',label='Dirty Price')
ax2.plot(times,dprices_5,linestyle='--',color='black')
ax3.plot(times,dprices_6,linestyle='--',color='black')


ax1.plot(times,cprices_4,linewidth=1,color='black',label='Clean Price')
ax2.plot(times,cprices_5,linewidth=1,color='black')
ax3.plot(times,cprices_6,linewidth=1,color='black')


ax1.yaxis.set_major_formatter(FuncFormatter(yfunc))
ax2.yaxis.set_major_formatter(FuncFormatter(yfunc))
ax3.yaxis.set_major_formatter(FuncFormatter(yfunc))

ax1.set_ylabel('Prix(P)',fontsize=15)
ax1.set_xlabel(u'Date de valeur: $t$ (années).',fontsize=15)
ax2.set_xlabel(u'Date de valeur: $t$ (années).',fontsize=15)
ax3.set_xlabel(u'Date de valeur: $t$ (années).',fontsize=15)

ax1.set_title(u'$R=$4\% et $C=$5\%',fontsize=17)
ax2.set_title(u'$R=$5\% et $C=$5\%',fontsize=17)
ax3.set_title(u'$R=$6\% et $C=$5\%',fontsize=17)


legend = ax1.legend(loc='lower left')

ax1.set_ylim(0.9,1.15)

plt.savefig('../figures/bondprices.png', bbox_inches='tight')
```
