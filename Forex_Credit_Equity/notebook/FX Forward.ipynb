```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
rcParams['text.latex.unicode']=True
```

```python
import numpy as np
import matplotlib.pyplot as plt


def plotFixFlow(ax,startx,t,value,color,label,size):
    ax.arrow(startx,t,value,0,color=color,head_width=0.04, head_length=0.02)
    if value<0:
        posx = 20
    else:
        posx = -120
    ax.annotate(label, xy=(startx,t),  
                xycoords='data',color = color, size=size, xytext=(posx, -10), textcoords='offset points'
                )

def figure(slide,filename):
    fig, ax = plt.subplots(figsize=(20,10))
    
    ax.set_xlim([0,1])
    ax.set_ylim([-0.5,0.5])
    
    plt.axis('off')

    ax.arrow(0.25,-0.4,0,0.8,color='black',head_width=0, head_length=0)
    ax.arrow(0.75,-0.4,0,0.8,color='black',head_width=0, head_length=0)
    
    ax.arrow(0.5,-0.4,0,0.8,color='black',head_width=0.02, head_length=0.04)
    
    ax.arrow(0.5,-0.4,0,0.8,color='black',head_width=0, head_length=0)
    
    ax.arrow(0.49,0.35,0.02,0,color='black',head_width=0, head_length=0,linewidth=0.05)
    
    ax.annotate('T', xy=(0.5,0.35),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    
    ax.arrow(0.49,-0.35,0.02,0,color='black',head_width=0, head_length=0)
    
    ax.annotate('t+2D', xy=(0.5,-0.35),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
    ax.arrow(0.49,-0.37,0.02,0,color='black',head_width=0, head_length=0)
    
    ax.annotate('t', xy=(0.5,-0.40),  
                    xycoords='data', size=25, xytext=(20, -10), textcoords='offset points'
                    )
        
    ax.annotate('USD', xy=(0.25,0.45),  
                    xycoords='data', size=25, xytext=(-20, 0), textcoords='offset points'
                    )
    
    ax.annotate('EUR', xy=(0.75,0.45),  
                    xycoords='data', size=25, xytext=(-20, 0), textcoords='offset points'
                    )
    
    ax.annotate('Temps', xy=(0.5,0.45),  
                    xycoords='data', size=25, xytext=(-40, 5), textcoords='offset points'
                    )
    
    if slide == 1:
        plotFixFlow(ax,0.75,0.35,0.1,'red','1 euro',25)
    else:
        plotFixFlow(ax,0.75,0.35,0.1,'red','',25)
        plotFixFlow(ax,0.75,0.35,-0.1,'red','',25)
        if slide == 2:
            plotFixFlow(ax,0.75,-0.35,0.06,'red','$\\frac{1}{1+\delta R^{EUR}}$',30)
        else:
            plotFixFlow(ax,0.75,-0.35,0.06,'red','',30)
            plotFixFlow(ax,0.75,-0.35,-0.06,'red','',30)
            if slide == 3:
                plotFixFlow(ax,0.25,-0.35,0.06,'blue','$\\frac{S}{1+\delta R^{EUR}}$',30)
            else:
                plotFixFlow(ax,0.25,-0.35,0.06,'blue','',25)
                plotFixFlow(ax,0.25,-0.35,-0.06,'blue','',25)
                plotFixFlow(ax,0.25,0.35,0.1,'blue','',25)
    
    if slide < 4:
        plotFixFlow(ax,0.25,0.35,-0.1,'blue','X dollar ?',25)
    else:
        plotFixFlow(ax,0.25,0.35,-0.1,'blue','',25)
        
    
    if slide >= 2:
        ax.annotate('=0', xy=(0.75,0.35),  
                    xycoords='data', size=25, xytext=(20, -30), textcoords='offset points'
                    )
    if slide >= 3:
        ax.annotate('=0', xy=(0.75,-0.35),  
                    xycoords='data', size=25, xytext=(20, 30), textcoords='offset points'
                    )
    if slide >= 4:
        ax.annotate('=0', xy=(0.25,-0.35),  
                    xycoords='data', size=25, xytext=(20, 30), textcoords='offset points'
                    )
        
    if slide >= 5:
        ax.annotate('=0', xy=(0.25,0.35),  
                    xycoords='data', size=25, xytext=(20, -30), textcoords='offset points'
                    )
    
    
    plt.savefig(filename, bbox_inches='tight')
    
```
