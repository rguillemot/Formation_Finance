```python
S0=1.1
rd=0.01
rf=0.005
T=1.0
vol=0.1
NT=12*4
N=1000
```

```python
dt = T/NT
```

```python
import numpy as np
import math
```

```python
np.random.seed(4)
```

```python
gaussians=np.random.randn(N,NT)
```

```python
spot = map(lambda y: map(lambda x: math.exp((rd-rf)*dt-0.5*vol*vol*dt+vol*sqrt(dt)*x),y),gaussians)
spot = map(lambda x: list(np.cumprod(x)),spot)
spot = np.array(spot)*S0
```

```python
Notional1 = 1000000.0
Notional2 = S0*Notional1
```

```python
Notional1*exp(-rf*T)*S0-Notional2*exp(-rd*T)
```

```python
timesteps = map(lambda x: float(x)/NT,range(1,NT+1))
```

```python
MTM = map(lambda x: map(lambda (y,t):Notional1*y*exp((rd-rf)*(T-t))*math.exp(-rd*T)-Notional2*math.exp(-rd*T),zip(x,timesteps)),spot)
```

```python
Notional1*math.exp((rd-rf)*T)*S0*math.exp(-rd*T)-Notional2*math.exp(-rd*T)
```

```python
MTM=np.array(MTM)
```

```python
np.std(map(lambda x: max(x,0),MTM.T[NT-1]))/np.sqrt(N)
```

```python
map(lambda x: np.mean(map(lambda x: x,x)),MTM.T)
```
