```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plotFlow(ax,t,value,color):
    ax.arrow(t,0,0,value,color=color,head_width=0.1, head_length=0.2,linewidth=2.5)
    


fig, ax = plt.subplots(figsize=(12,6))

ax.set_xlim([-0.8,3])
ax.set_ylim([-1.7,1.7])

x = np.arange(0.5,2+0.1,0.1)
y = -(x-0.5)*(x-2)*0.3

ax.plot(x,y,color='black',linewidth=0.2,linestyle='dashed')

ax.annotate('$\\delta=\\frac{T_{end}-T_{start}}{360}$', xy=(1,0.4),  xycoords='data', size=20,
                xytext=(0, 0), textcoords='offset points'
            )


plt.axis('off')

ax.arrow(-0.5,0,3,0,color='black',head_width=0.2, head_length=0.1,linewidth=0.1)
ax.arrow(0,-0.1,0,0.2,color='black',linewidth=0.1)

ax.arrow(0.4,-0.1,0,0.2,color='black',linewidth=0.1)

plotFlow(ax,0.5,-1,'black')
plotFlow(ax,2,1.2,'black')

ax.annotate('$t$', xy=(0,0),  xycoords='data', size=20,
                xytext=(10, 20), textcoords='offset points'
                )

ax.annotate('$T_{start}$', xy=(0.5,0),  xycoords='data', size=20,
                xytext=(10, 20), textcoords='offset points'
                )

ax.annotate('$T_{end}$', xy=(2,0),  xycoords='data', size=20,
                xytext=(10, 20), textcoords='offset points'
                )

ax.annotate('$1+\delta K$', xy=(2,0.7),  xycoords='data', size=20,
                xytext=(10, 20), textcoords='offset points'
                )

ax.annotate('$1$', xy=(0.5,-0.7),  xycoords='data', size=20,
                xytext=(10, 20), textcoords='offset points'
                )

ax.annotate('$T_f$', xy=(0.4,0),  xycoords='data', size=20,
                xytext=(-20, -30), textcoords='offset points'
                )

plt.savefig('../figures/fwd_replic1.png', bbox_inches='tight')
```
