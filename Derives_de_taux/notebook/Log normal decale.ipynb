```python
import numpy as np
from scipy.stats import norm
import math

def BS(F,K,T,cp,vol):
    if vol*T == 0:
        return cp*(F-K)
    else:
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        d2 = d1 - vol*math.sqrt(T)
        return cp*F*norm.cdf(cp*d1)-cp*K*norm.cdf(cp*d2)
    
def ImpliedVol(F,K,T,cp,price):
    vol = 0.1
    eps = 1e-8
    for i in range(20):
        val = BS(F,K,T,cp,vol)
        if abs(val-price) < eps:
            return vol
        deriv = (BS(F,K,T,cp,vol+eps)-val)/eps
        vol = vol -(val-price)/deriv
    return vol
        
def ImpliedVolSL(F,K,T,cp,vol,m):
    return ImpliedVol(F,K,T,cp,BS(F+m,K+m,T,cp,vol))
```

```python
import matplotlib.pyplot as plt
F=0.05
K=0.05
T=1.0
vol = 0.18
ms = [0,0.02,0.1,1]
strikes = np.arange(0.02,0.082,0.002)
cps = [1 if k > F else -1 for k in strikes ]
fig, ax = plt.subplots(figsize=(12,6))
for m in ms: 
    vols = [ImpliedVolSL(F,k,T,acp,vol*F/(F+m),m) for k,acp in zip(strikes,cps)]
    ax.plot(strikes,vols)
```

```python
from scipy.stats import lognorm

volsSLN = [vol*F/(F+m) for m in ms]

fig, ax = plt.subplots(figsize=(12,6))
for m,vol in zip(ms,volsSLN):
    x = np.arange(-m+0,-m+1.1,0.1)
    y = [lognorm.pdf((val+m)/vol) for val in x]
    ax.plot(x,y)
y
```
