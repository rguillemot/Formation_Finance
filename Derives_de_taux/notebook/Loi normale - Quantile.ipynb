```python
from scipy.stats import norm
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(norm.ppf(0.001),norm.ppf(0.999), 100)
ax.set_ylim(-0.1,0.5)
ax.set_xlim(-2.6,2.6)
ax.spines['right'].set_color('none')
ax.spines['right'].set_visible(False)
ax.spines['left'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()


ax.annotate('', xy=(-1,-0.1),  xycoords='data', size=40,
                xytext=(0, 220), textcoords='offset points',
                arrowprops=dict(arrowstyle="-",linestyle='dashed'),
                )

ax.annotate('', xy=(1,-0.1),  xycoords='data', size=40,
                xytext=(0, 220), textcoords='offset points',
                arrowprops=dict(arrowstyle="-",linestyle='dashed'),
                )

ax.set_xticks([])
ax.set_yticks([])

ax.plot(x, norm.pdf(x),'blue', alpha=0.6, label='norm pdf')
fig.show()
```

```python
"""
Plot demonstrating the integral as the area under a curve.

Although this is a simple example, it demonstrates some important tweaks:

    * A simple line plot with custom color and line width.
    * A shaded region created using a Polygon patch.
    * A text label with mathtext rendering.
    * figtext calls to label the x- and y-axes.
    * Use of axis spines to hide the top and right spines.
    * Custom tick placement and labels.
"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon


def func(x):
    return norm.pdf(x)


x = np.linspace(-3, 3)
y = func(x)

fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(15,4))

def plot_quantile(ax,q,label):
    ax.plot(x, y, 'black', linewidth=2)
    #plt.ylim(ymin=0)

    # Make the shaded region
    ix = np.linspace(-q, q)
    iy = func(ix)
    verts = [(-q, 0)] + list(zip(ix, iy)) + [(q, 0)]
    poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
    ax.add_patch(poly)

    ax.text(0.5 * 0, 0.17, r"$Pr(|G|<"+str(q)+r")$",
             horizontalalignment='center', fontsize=18)

    ax.text(0.5 * 0, 0.13, r"="+label,
             horizontalalignment='center', fontsize=18)


    #plt.figtext(0.9, 0.05, '$x$')
    #plt.figtext(0.1, 0.9, '$y$')

    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')

    ax.set_xticks((-q, q))
    ax.set_xticklabels((-q, q),fontsize=16)
    ax.set_yticks([])

    ax.set_title(r'Quantile '+label,fontsize=16)
    
plot_quantile(ax[0],1,r'$68\%$')
plot_quantile(ax[1],1.96,r'$95\%$')

fig.suptitle("Loi Normale", fontsize=20)

plt.savefig('../figures/loinormale_quantile.png', bbox_inches='tight')
```
