```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
from scipy.stats import lognorm
from scipy.stats import norm
```

```python
def myfunc(x, pos=0):
     return r'%1.0f'%(100*x)+r'\%'
```

```python
T = 10
F1 = 0.05
vol_N1 = 0.01
vol_LN1 = vol_N1/F1
x_LN1 = np.linspace(0.02,0.1, 100)
x_N1 = np.linspace(0.02,0.1, 100)
x_N1_vol = np.linspace(0.04,0.06, 100)
y_LN1 = norm.pdf(log(x_LN1/F1)/vol_LN1)
y_LN1_cum = norm.cdf(log(x_LN1/F1)/vol_LN1)
y_N1 = norm.pdf((x_N1-F1)/vol_N1)
y_N1_cum = norm.cdf((x_N1-F1)/vol_N1)

F2 = 0.01
vol_N2 = 0.01
vol_LN2 = vol_N2/F2
x_LN2 = np.linspace(0.0001,0.08, 100)
x_N2 = np.linspace(-0.0200,0.08, 100)
x_N2_vol = np.linspace(0.005,0.02, 100)
y_LN2 = norm.pdf(log(x_LN2/F2)/vol_LN2)
y_LN2_cum = norm.cdf(log(x_LN2/F2)/vol_LN2)
y_N2 = norm.pdf((x_N2-F2)/vol_N2)
y_N2_cum = norm.cdf((x_N2-F2)/vol_N2)
```

```python
def BS(F,K,T,cp,vol):
    if vol*T == 0:
        return cp*(F-K)
    else:
        d1 = math.log(F/K)/vol/math.sqrt(T)+vol*math.sqrt(T)/2.0
        d2 = d1 - vol*math.sqrt(T)
        return cp*F*norm.cdf(cp*d1)-cp*K*norm.cdf(cp*d2)
    
def Bachelier(F,K,T,cp,vol):
    if vol*T == 0:
        return cp*(F-K)
    else:
        d = cp*(F-K)/vol/math.sqrt(T)
        return vol*sqrt(T)*(d*norm.cdf(d)+norm.pdf(d))
    
def ImpliedVol(F,K,T,cp,price):
    vol = 0.1
    eps = 1e-8
    for i in range(20):
        val = BS(F,K,T,cp,vol)
        if abs(val-price) < eps:
            return vol
        deriv = (BS(F,K,T,cp,vol+eps)-val)/eps
        vol = vol -(val-price)/deriv
    return vol
```

```python
smile_LN1 = np.repeat(vol_LN1,100)
smile_N1 = map(lambda x: ImpliedVol(F1,x,1,1,Bacherlier(F1,x,1,1,vol_N1)),x_N1_vol)
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(15,10))

ax2 = ax[0][0].twinx()
ax[0][0].plot(x_LN1,y_LN1,color='blue',label='LN')
ax2.plot(x_LN1,y_LN1_cum,color='blue',ls='--')
ax[0][0].plot(x_N1,y_N1,color='red',label='Gauss')
ax2.plot(x_LN1,y_N1_cum,color='red',ls='--')
ax[0][0].legend(loc='best', frameon=False)
ax[0][0].xaxis.set_major_formatter(FuncFormatter(myfunc))
ax[0][0].yaxis.set_major_formatter(FuncFormatter(myfunc))
ax[0][0].set_ylabel('densite',fontsize=15)
ax[0][0].set_xlabel('taux',fontsize=15)
ax2.set_ylabel('repartition',fontsize=15)
ax2.xaxis.set_major_formatter(FuncFormatter(myfunc))
ax2.yaxis.set_major_formatter(FuncFormatter(myfunc))

ax[0][0].set_title('F=5\% $\sigma^{LN}$=20\% $\sigma^{N}$=1\%',fontsize=16)

ax2 = ax[0][1].twinx()
ax[0][1].plot(x_LN2,y_LN2,color='blue',label='LN')
ax2.plot(x_LN2,y_LN2_cum,color='blue',ls='--')
ax[0][1].plot(x_N2,y_N2,color='red',label='Gauss')
ax2.plot(x_N2,y_N2_cum,color='red',ls='--')

ax[0][1].legend(loc='best', frameon=False)
ax[0][1].set_xlabel('taux',fontsize=15)
ax[0][1].xaxis.set_major_formatter(FuncFormatter(myfunc))
ax[0][1].yaxis.set_major_formatter(FuncFormatter(myfunc))
ax2.xaxis.set_major_formatter(FuncFormatter(myfunc))
ax2.yaxis.set_major_formatter(FuncFormatter(myfunc))

ax[0][1].set_title('F=1\% $\sigma^{LN}$=100\% $\sigma^{N}$=1\%',fontsize=16)

ax[1][0].plot(x_N1_vol,smile_LN1,color='blue',label='LN')
ax[1][0].plot(x_N1_vol,smile_N1,color='red',label='Gauss')
ax[1][0].legend(loc='best', frameon=False)
ax[1][0].xaxis.set_major_formatter(FuncFormatter(myfunc))
ax[1][0].yaxis.set_major_formatter(FuncFormatter(myfunc))
ax[1][0].set_ylabel('smile',fontsize=15)
ax[1][0].set_xlabel('taux',fontsize=15)

ax[1][0].set_xlim([0.02,0.1])
ax[1][0].set_ylim([0.17,0.23])

ax[1][1].spines['top'].set_visible(False)
ax[1][1].spines['bottom'].set_visible(False)
ax[1][1].spines['left'].set_visible(False)
ax[1][1].spines['right'].set_visible(False)
ax[1][1].set_xticks([])
ax[1][1].set_yticks([])

plt.savefig('../figures/LognormalvsNormal.png', bbox_inches='tight')
```
