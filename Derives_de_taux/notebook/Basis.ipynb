```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plotFixFlow(ax,starty,t,value,color,label):
    ax.arrow(t,starty,0,value,color=color,head_width=0.05, head_length=0.1,linewidth=0.4)
    ax.annotate(label, xy=(t,starty),  
                xycoords='data', size=20, xytext=(10, -25), textcoords='offset points'
                )


def plotVarFlow(ax,starty,t,value,color,label):
    y = starty + np.arange(0,value+0.005,0.005)
    x = [t+np.sin(v/value*np.pi*10)*0.02 for v in y]
    ax.plot(x,y,color='black',linewidth=0.4)
    ax.arrow(t,starty+value+0.02,0,0.0001,color=color,head_width=0.05, head_length=0.1,linewidth=2.5)
    ax.annotate(label, xy=(t,starty),  
                xycoords='data', size=20, xytext=(10, -25), textcoords='offset points'
                )
    
def plotdelta(ax,starty,t1,t2,color,label):
    x = np.arange(t1,t2+0.02,0.02)
    y = starty-(x-t1)*(x-t2)/(t1-t2)/(t1-t2)*0.3
    ax.plot(x,y,color=color,linewidth=0.4,linestyle='--')
    ax.annotate(label, xy=((t1+t2)/2,starty),  
                xycoords='data', size=20, xytext=(-10, 10), textcoords='offset points'
                )
    
def plotdash(y,t1,t2):
    x = np.arange(t1,t2+0.02,0.02)
    y = [y for i in range(len(x))]
    ax.plot(x,y,color='black',linewidth=0.4,linestyle='--')



fig, ax = plt.subplots(figsize=(15,7.5))

ax.set_xlim([-0.8,3])
ax.set_ylim([-1.2,2])

plt.axis('off')

ax.arrow(0,0.9,0,0.2,color='black',linewidth=0.1)
ax.annotate('$T^m_0=t$', xy=(0,1),  
                xycoords='data', size=20, xytext=(-30, -25), textcoords='offset points'
                )
ax.arrow(0,-0.1,0,0.2,color='black',linewidth=0.1)
ax.annotate('$T^n_0=t$', xy=(0,0),  
                xycoords='data', size=20, xytext=(-30, -25), textcoords='offset points'
                )
ax.arrow(0,-1.1,0,0.2,color='black',linewidth=0.1)
ax.annotate('$T_0=t$', xy=(0,-1),  
                xycoords='data', size=20, xytext=(-30, -25), textcoords='offset points'
                )

ax.arrow(-0.5,1,3,0,color='black',head_width=0.2, head_length=0.1,linewidth=0.1)

ax.arrow(-0.5,0,3,0,color='black',head_width=0.2, head_length=0.1,linewidth=0.1)

ax.arrow(-0.5,-1,3,0,color='black',head_width=0.2, head_length=0.1,linewidth=0.1)

plotdash(1.3,1,1.8)
plotdash(0.3,1,1.8)
plotdash(-0.7,1,1.8)

plotVarFlow(ax,1,0.2,0.4,'black','$T^n_1$')
plotVarFlow(ax,1,0.4,0.4,'black','$T^n_2$')
plotVarFlow(ax,1,0.6,0.4,'black','$T^n_3$')
plotVarFlow(ax,1,0.8,0.4,'black','$T^n_4$')
plotVarFlow(ax,1,2.0,0.4,'black','$T^n_n$')

plotdelta(ax,1,0,0.2,'black','$\delta^n_1$')
plotdelta(ax,1,0.2,0.4,'black','$\delta^n_2$')

plotVarFlow(ax,0,0.4,0.4,'black','$T^m_1$')
plotVarFlow(ax,0,0.8,0.4,'black','$T^m_2$')
plotVarFlow(ax,0,2.0,0.4,'black','$T^m_m$')

plotdelta(ax,0,0,0.4,'black','$\delta^m_1$')
plotdelta(ax,0,0.4,0.8,'black','$\delta^m_2$')

plotFixFlow(ax,-1,0.4,0.4,'black','$T_1$')
plotFixFlow(ax,-1,0.8,0.4,'black','$T_2$')
plotFixFlow(ax,-1,2.0,0.4,'black','$T_n$')

plotdelta(ax,-1,0,0.4,'black','$\delta_1$')
plotdelta(ax,-1,0.4,0.8,'black','$\delta_2$')

ax.annotate('Jambe variable a n flux.', xy=(-0.8,1.2),  
                xycoords='data', size=17, xytext=(0, 10), textcoords='offset points'
                )

ax.annotate('Jambe variable a m flux.', xy=(-0.8,0.2),  
                xycoords='data', size=17, xytext=(0, 10), textcoords='offset points'
                )

ax.annotate('Jambe fixe.', xy=(-0.8,-0.8),  
                xycoords='data', size=17, xytext=(0, 10), textcoords='offset points'
                )

plt.savefig('../figures/basis.png', bbox_inches='tight')
```
